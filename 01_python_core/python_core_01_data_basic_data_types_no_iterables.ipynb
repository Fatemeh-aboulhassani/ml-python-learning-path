## 1.1 What is data?

Data is information.

A program receives data, stores data, processes data, and produces a result.

Examples of data:

- a user's name
- a user's age
- a product price
- a temperature value
- a model accuracy value
- a robot battery level
- a login status
- a file path
- a message
- an error status

A simple program usually works like this:

```text
input data -> processing -> output result
```

Example:

```text
price + quantity -> calculate total price -> print invoice
```

In Python, every piece of data has a **value** and a **type**.

Example:

```python
25
```

Value: `25`  
Type: integer number

Example:

```python
"Python"
```

Value: `"Python"`  
Type: text/string

````python
# Basic data examples

print(25)
print(19.99)
print("Python")
print(True)
print(None)
````

## 1.2 What is a variable?

A variable is a name that stores a value.

Example:

```python
age = 25
```

This means:

- `age` is the variable name
- `25` is the value
- `=` assigns the value to the variable

Important:

In Python, `=` does not mean mathematical equality.  
It means assignment.

```python
name = "Ali"
```

This means: store `"Ali"` inside the variable `name`.

````python
age = 25
name = "Ali"
price = 19.99
is_active = True

print(age)
print(name)
print(price)
print(is_active)
````

## 1.3 Variable naming rules

A variable name:

- can contain letters
- can contain numbers
- can contain underscores
- cannot start with a number
- cannot contain spaces
- cannot contain hyphens
- cannot be a Python reserved keyword

Good variable names:

```python
user_name
user_age
product_price
model_accuracy
robot_battery_level
```

Bad variable names:

```python
2user
user-name
user name
class
```

Professional rule:

Use meaningful names.

Bad:

```python
x = 0.93
```

Good:

```python
model_accuracy = 0.93
```

Python style normally uses `snake_case`.

````python
user_name = "Sara"
user_age = 28
product_price = 49.99
model_accuracy = 0.93
robot_battery_level = 87

print(user_name)
print(user_age)
print(product_price)
print(model_accuracy)
print(robot_battery_level)
````

## 1.4 What is a data type?

A data type tells Python what kind of value it is.

The data type controls what operations are possible.

For example:

- numbers can be added
- strings can be joined
- booleans can represent yes/no states
- `None` can represent missing data

In this notebook, we study these basic non-iterable data types:

| Type | Meaning | Example |
|---|---|---|
| `int` | whole number | `25` |
| `float` | decimal number | `19.99` |
| `str` | text | `"Python"` |
| `bool` | true or false | `True` |
| `NoneType` | no value | `None` |

````python
age = 25
price = 19.99
language = "Python"
is_online = True
last_error = None

print(type(age))
print(type(price))
print(type(language))
print(type(is_online))
print(type(last_error))
````

## 1.5 Integer: `int`

An integer is a whole number.

Examples:

```python
10
0
-5
2026
```

Common uses:

- age
- quantity
- number of users
- number of attempts
- number of training epochs
- status code
- ID number

````python
user_age = 25
product_quantity = 3
http_status_code = 200
training_epochs = 10
failed_login_attempts = 0

print(user_age)
print(product_quantity)
print(http_status_code)
print(training_epochs)
print(failed_login_attempts)

print(type(user_age))
````

Common operations with integers:

````python
a = 10
b = 3

print(a + b)  # addition
print(a - b)  # subtraction
print(a * b)  # multiplication
print(a / b)  # division
print(a // b) # floor division
print(a % b)  # remainder
print(a ** b) # power
````

Important note:

Normal division `/` always returns a `float`, even when the result looks like a whole number.

````python
result = 10 / 2

print(result)
print(type(result))
````

## 1.6 Float: `float`

A float is a decimal number.

Examples:

```python
19.99
0.95
-3.14
100.0
```

Common uses:

- price
- temperature
- model accuracy
- model loss
- sensor value
- GPS coordinate
- distance
- speed

````python
product_price = 49.99
temperature = 24.7
model_accuracy = 0.9345
sensor_distance = 1.75
robot_speed = 0.8

print(product_price)
print(temperature)
print(model_accuracy)
print(sensor_distance)
print(robot_speed)

print(type(model_accuracy))
````

Floats are useful, but they can sometimes have small precision behavior because computers store decimal numbers in binary form.

````python
result = 0.1 + 0.2

print(result)
````

This is normal in many programming languages.

For most beginner Python work, this is not a problem.

For financial systems, you should later learn about `Decimal`.

## 1.7 String: `str`

A string is text.

You can create strings with single quotes or double quotes.

```python
"Python"
'Python'
```

Common uses:

- name
- email
- password
- file path
- message
- API token
- model name
- command text

````python
first_name = "Ali"
email = "ali@example.com"
model_name = "Linear Regression"
file_path = "data/train.csv"
message = "Welcome to Python Core"

print(first_name)
print(email)
print(model_name)
print(file_path)
print(message)

print(type(message))
````

### Multi-line strings

A multi-line string can be created with triple quotes.

````python
description = """
This is a simple Python notebook.
It teaches basic data types.
It does not cover iterable data types yet.
"""

print(description)
````

### String concatenation

Concatenation means joining strings.

````python
first_name = "Ali"
last_name = "Ahmadi"

full_name = first_name + " " + last_name

print(full_name)
````

### String repetition

````python
line = "-" * 20

print(line)
print("Report")
print(line)
````

### Useful string methods

A method is an action that belongs to a value.

For strings, common methods include:

- `.lower()`
- `.upper()`
- `.strip()`
- `.replace()`
- `.startswith()`
- `.endswith()`

````python
text = "  Python Core  "

print(text.lower())
print(text.upper())
print(text.strip())
print(text.replace("Core", "Basics"))
print(text.strip().startswith("Python"))
print(text.strip().endswith("Core"))
````

## 1.8 Boolean: `bool`

A boolean has only two possible values:

```python
True
False
```

Important:

- `True` must start with capital `T`
- `False` must start with capital `F`

Common uses:

- is user logged in?
- is payment successful?
- is model trained?
- is robot moving?
- is device online?
- is file available?

````python
is_logged_in = True
is_payment_successful = False
is_model_trained = True
is_robot_moving = False
is_device_online = True

print(is_logged_in)
print(is_payment_successful)
print(is_model_trained)
print(is_robot_moving)
print(is_device_online)

print(type(is_logged_in))
````

Booleans are often created from comparisons.

````python
age = 20
is_adult = age >= 18

print(is_adult)
print(type(is_adult))
````

## 1.9 None: missing or empty value

`None` means no value.

It is not the same as:

- `0`
- empty string `""`
- `False`

`None` means the value does not exist yet or is unknown.

Common uses:

- no error yet
- no result yet
- no user selected yet
- no sensor data yet
- no prediction yet

````python
last_error = None
prediction_result = None
selected_user = None
sensor_value = None

print(last_error)
print(prediction_result)
print(selected_user)
print(sensor_value)

print(type(last_error))
````

Checking if something is None:

````python
last_error = None

print(last_error is None)
print(last_error is not None)
````

## 1.10 Type checking with `type()`

Use `type()` to check the exact type of a value.

````python
value_1 = 100
value_2 = 19.99
value_3 = "Python"
value_4 = True
value_5 = None

print(type(value_1))
print(type(value_2))
print(type(value_3))
print(type(value_4))
print(type(value_5))
````

## 1.11 Type checking with `isinstance()`

Use `isinstance()` when you want to check if a value belongs to a type.

This is common in real projects.

````python
age = 25
price = 19.99
name = "Sara"
is_active = True

print(isinstance(age, int))
print(isinstance(price, float))
print(isinstance(name, str))
print(isinstance(is_active, bool))
````

Important note:

In Python, `bool` is related to `int`.

So this can be surprising:

````python
print(isinstance(True, bool))
print(isinstance(True, int))
````

For beginner work, remember:

- use `bool` for logical states
- use `int` for whole numbers
- do not use `True` or `False` as numbers in normal application code

## 1.12 Type conversion

Type conversion means changing a value from one type to another.

Common conversion functions:

- `int()`
- `float()`
- `str()`
- `bool()`

````python
age_text = "25"
age_number = int(age_text)

print(age_number)
print(type(age_number))
````

````python
price_text = "49.99"
price_number = float(price_text)

print(price_number)
print(type(price_number))
````

````python
score = 98
score_text = str(score)

print(score_text)
print(type(score_text))
````

### Be careful with invalid conversions

This will fail:

```python
int("twenty")
```

because `"twenty"` is text, not a numeric string.

````python
# Uncomment to see the error:
# age = int("twenty")
````

## 1.13 Truthy and falsy values

Some values behave like `False` when converted to bool.

Examples:

- `0`
- `0.0`
- `""`
- `None`

Many other values behave like `True`.

````python
print(bool(0))
print(bool(0.0))
print(bool(""))
print(bool(None))

print(bool(1))
print(bool(-5))
print(bool("Python"))
````

## 1.14 Input with `input()`

`input()` gets data from the user.

Very important:

`input()` always returns a string.

````python
name = input("Enter your name: ")

print(name)
print(type(name))
````

````python
age = input("Enter your age: ")

print(age)
print(type(age))
````

If you need a number, convert it.

````python
age = int(input("Enter your age: "))

print(age)
print(type(age))
````

````python
price = float(input("Enter product price: "))

print(price)
print(type(price))
````

## 1.15 Output with `print()`

`print()` shows output to the user.

````python
name = "Ali"
age = 25

print("Name:", name)
print("Age:", age)
````

## 1.16 f-strings

An f-string is a clean way to put variables inside text.

This is the professional style for simple output.

````python
name = "Ali"
age = 25
city = "Berlin"

print(f"Name: {name}")
print(f"Age: {age}")
print(f"City: {city}")
````

## 1.17 Number formatting

You can control how numbers appear in output.

Useful formatting:

```python
{value:.2f}
```

This means: show 2 digits after the decimal point.

````python
price = 49.9999
accuracy = 0.934567
temperature = 24.789

print(f"Price: ${price:.2f}")
print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Temperature: {temperature:.1f}°C")
````

## 1.18 Multiple assignment

You can assign multiple variables in one line.

````python
name, age, city = "Sara", 28, "Tehran"

print(name)
print(age)
print(city)
````

Use this only when it improves readability.

## 1.19 Dynamic typing

Python is dynamically typed.

This means a variable can point to different types at different times.

````python
value = 100
print(value)
print(type(value))

value = "one hundred"
print(value)
print(type(value))
````

This is allowed, but it can make code confusing.

Professional code should avoid changing the meaning and type of a variable without a good reason.

Better:

```python
user_score = 100
score_text = "one hundred"
```

## 1.20 Common beginner mistakes

### Mistake 1: adding string and number

````python
# Uncomment to see the error:
# age = 25
# message = "Age: " + age
````

Correct:

````python
age = 25

message = "Age: " + str(age)
print(message)

print(f"Age: {age}")
````

### Mistake 2: forgetting that input returns string

````python
x = "10"
y = "20"

print(x + y)
````

Correct:

````python
x = int("10")
y = int("20")

print(x + y)
````

### Mistake 3: unclear variable names

Bad:

```python
a = 49.99
b = 3
c = a * b
```

Better:

```python
unit_price = 49.99
quantity = 3
total_price = unit_price * quantity
```

## 1.21 Real-world example: product price report

````python
product_name = "Wireless Mouse"
unit_price = 29.99
quantity = 3
discount_percent = 10.0

subtotal = unit_price * quantity
discount_amount = subtotal * discount_percent / 100
final_price = subtotal - discount_amount

print("Product Price Report")
print("--------------------")
print(f"Product: {product_name}")
print(f"Unit Price: ${unit_price:.2f}")
print(f"Quantity: {quantity}")
print(f"Subtotal: ${subtotal:.2f}")
print(f"Discount: ${discount_amount:.2f}")
print(f"Final Price: ${final_price:.2f}")
````
